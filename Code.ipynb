```python
import pandas as pd
data = pd.read_csv('data/contients_population.csv')
data1=data
data1
```

```python
from scipy.stats import stats
def getP_Value(list1,list2):
    res = stats.ttest_ind(list1, list2, 
                      equal_var=True)
    display(res)
```

```python
def data_via (val,col):
    result_Europe=data.loc[data['Area']==val]
    lst_Europe=[]
    print(val)
    mean_Europe=result_Europe.mean()
    median_Europe=result_Europe.median()
    std_Europe=result_Europe.std()
    lst_Europe.append(mean_Europe)
    lst_Europe.append(median_Europe)
    lst_Europe.append(std_Europe)
    df = pd.DataFrame(lst_Europe,columns=[col])
    df.index=['Mean','Median','Std']
    print(df)
    print("\n")
```

```python
data_via('Europe','Density (P/KmÂ²)')
```

```python
data_via('Africal','Density (P/KmÂ²)')
```

```python
def Boxplot(data,res,ylabel,name,ymin,ymax):
    dt_africal = data.loc[data['Area']=='Africal']
    dt_asian = data.loc[data['Area']=='Asian']
    dt_europe = data.loc[data['Area']=='Europe']
    dt_north_american = data.loc[data['Area']=='North American']
    dt_south_american = data.loc[data['Area']=='South American']
    data1 = [dt_africal[res], dt_asian[res], dt_europe[res], dt_north_american[res], dt_south_american[res]]
    fig, (ax1) = plt.subplots(nrows = 1, ncols = 1, figsize = (10,10))
    ax1.boxplot(data1, notch=False, meanline=True,showmeans=True,widths=0.3,patch_artist = True );
    ax1.grid(True)
    ax1.set_xticklabels(["Africal", "Asian", "Europe", "North American", "South American"], fontsize = 10, fontweight = "bold")
    fig.subplots_adjust(
        hspace = 0.15,
        top = 0.94,
        right = 0.9
    );
    ax1.set_ylim(ymin,ymax);
    fig.suptitle(name,
                 fontsize = 12, color = "green", fontweight = "bold")
    ax1.set_xlabel("AREA", fontsize = 14, fontstyle = "italic", color = "red")
    ax1.set_ylabel(ylabel, fontsize = 14, fontstyle = "italic", color = "red")
```

```python
name = 'Đồ thị boxplot thể hiện mật độ dân số của các khu vực trên thế giới'
Boxplot(data,'Density (P/KmÂ²)','Density',name,0,800)
```

```python
import matplotlib.pyplot as plt
import statistics
from math import sqrt
def plot_confidence_interval(x, values, z=1.96, color='#2187bb', horizontal_line_width=0.25):
    mean = statistics.mean(values)
    stdev = statistics.stdev(values)
    confidence_interval = z * stdev / sqrt(len(values))

    left = x - horizontal_line_width / 2
    top = mean - confidence_interval
    right = x + horizontal_line_width / 2
    bottom = mean + confidence_interval
    plt.plot([x, x], [top, bottom], color=color)
    plt.plot([left, right], [top, top], color=color)
    plt.plot([left, right], [bottom, bottom], color=color)
    plt.plot(x, mean, 'o', color='#f44336')

    return mean, confidence_interval
```

```python
def data_resl (Area1,Area2,col):
    result_val1=data.loc[data['Area']==Area1]
    result_val2=data.loc[data['Area']==Area2]

    lst_val1=result_val1[col]
    lst_val1=lst_val1.tolist()

    lst_val2=result_val2[col]
    lst_val2=lst_val2.tolist()
    return lst_val1,lst_val2
```

```python
lst_europe_Density,lst_africa_Density=data_resl('Europe','Africal','Density (P/KmÂ²)')
```

```python
plt.xticks([1, 2], ['Europe', 'Africal'])
plt.title('Confidence Interval')
plot_confidence_interval(1, lst_europe_Density)
plot_confidence_interval(2, lst_africa_Density)
plt.show()
```

```python
getP_Value(lst_europe_Density,lst_africa_Density)
```
